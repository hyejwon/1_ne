## 경계에 걸린 패치 제거하기

```python
def get_del_list(img, thr, minmax = 'min', lb='l'): #comp = true
    n_zero = 0
    del_list = []
    
    
    for i in range(img.shape[0]):
        if (minmax=='min'):
            v = img[i].min()
        elif (minmax=='max'):
            v = img[i].max()
        elif (minmax=='mean'):
            v = img[i].mean()
        else:
            print('error')
           
        if(v < thr and lb=='l'): 
            n_zero = n_zero +1
            del_list.append(i)
            
        elif(v > thr and lb=='u'): 
            n_zero = n_zero +1
            del_list.append(i)  
            
    print(n_zero)
    
    
    return del_list

# def delete_data_list()
def delete_data_list(del_list, traintest = 'train'):
    global low_img_train, high_img_train, disp_train, se_train
    global low_img_val, high_img_val, disp_val, se_val
    global low_img_test, high_img_test, disp_test, se_test
    if traintest =='train':
        low_img_train = np.delete(low_img_train, del_list,axis=0)
        high_img_train = np.delete(high_img_train, del_list,axis=0)
        disp_train = np.delete(disp_train, del_list,axis=0)
        se_train = np.delete(se_train, del_list,axis=0)

    if traintest =='val':
        low_img_val = np.delete(low_img_val, del_list,axis=0)
        high_img_val = np.delete(high_img_val, del_list,axis=0)
        disp_val = np.delete(disp_val, del_list,axis=0)
        se_val = np.delete(se_val, del_list,axis=0)
        
    if traintest =='test':
        low_img_test = np.delete(low_img_test, del_list,axis=0)
        high_img_test = np.delete(high_img_test, del_list,axis=0)
        disp_test = np.delete(disp_test, del_list,axis=0)
        se_test = np.delete(se_test, del_list,axis=0)
        
def del_abnormal_img(thr, minmax = 'min', lb='l'):
    del_list = get_del_list(high_img_train,thr,minmax,lb)
    delete_data_list(del_list, traintest='train')
    
    del_list = get_del_list(high_img_val,thr,minmax,lb)
    delete_data_list(del_list, traintest='val')
    
    del_list = get_del_list(high_img_test,thr,minmax,lb)
    delete_data_list(del_list, traintest='test')    
```

```python
## 경계에 걸린놈 제거하기. 빈공간의 경우 displacement 가 0로 세팅되어 있기 때문에 미분계산시 오류 발생. 

del_abnormal_img(0.0001,minmax = 'min',lb='l')
```

## 데이터를 augmentation 하기위한 함수 정의 (90도씩 돌리고 flip -> 8배)

```python
from scipy.ndimage import rotate
import math

# plt.imshow(low[500].reshape(img_rows, img_cols), cmap='Greys', vmin=0, vmax=1)
# plt.show()

def rot_augment(img):
    img90 = np.rot90(img,k=1,axes=(2,1))
    img180 = np.rot90(img,k=2,axes=(2,1))
    img270 = np.rot90(img,k=3,axes=(2,1))

    return np.concatenate((img, img90, img180, img270), axis=0)
    # return img

def flip_augment(img):
    fl=img[:,:,::-1,:]

#     fl=img[:,::-1,:,:]

    return np.concatenate((img,fl),axis=0)

def flip_augment_vector(img):
#     fl=-img[:,::-1,:,:]
    fl=-img[:,:,::-1,:]
    return np.concatenate((img,fl),axis=0)

def augment(img):
    return flip_augment(rot_augment(img))
    #return rot_augment(img)

def rot_vector(vx,vy,k):
    vx = np.rot90(vx,k,axes=(2,1))
    vy = np.rot90(vy,k,axes=(2,1))

    th = (2*3.141592)*(k*90/360)
    print('th=',th)
    vx_rot = vx*math.cos(th)-vy*math.sin(th)
    vy_rot = vx*math.sin(th)+vy*math.cos(th)
    
    return vx_rot, vy_rot

def rot_augment_vector(vx, vy):

    vx90, vy90 = rot_vector(vx,vy,1)
    vx180, vy180 = rot_vector(vx,vy,2)
    vx270, vy270 = rot_vector(vx,vy,3)

    vx = np.concatenate((vx,vx90,vx180,vx270),axis=0)
    vy = np.concatenate((vy,vy90,vy180,vy270),axis=0)
    
    return flip_augment_vector(vx), flip_augment(vy)

def rot_augment_vector2(v):

    vx = v[:,:,:,:3]
    vy = v[:,:,:,3:]
    
    vx_arg, vy_arg = rot_augment_vector(vx,vy)
    return np.concatenate((vx_arg,vy_arg),axis=3)


#return vx, vy
    


```

```python
def SubpixelConv2D(input_shape, scale=4):
    """
    Keras layer to do subpixel convolution.
    NOTE: Tensorflow backend only. Uses tf.depth_to_space
    Ref:
        [1] Real-Time Single Image and Video Super-Resolution Using an Efficient Sub-Pixel Convolutional Neural Network
            Shi et Al.
            https://arxiv.org/abs/1609.05158
    :param input_shape: tensor shape, (batch, height, width, channel)
    :param scale: upsampling scale. Default=4
    :return:
    """
    # upsample using depth_to_space
    def subpixel_shape(input_shape):
        dims = [input_shape[0],
                input_shape[1] * scale,
                input_shape[2] * scale,
                int(input_shape[3] / (scale ** 2))]
        output_shape = tuple(dims)
        return output_shape

    def subpixel(x):
        return tf.nn.depth_to_space(x, scale)


#     return Lambda(subpixel, output_shape=subpixel_shape, name='subpixel')
    return Lambda(subpixel, output_shape=subpixel_shape)

```

```python
def custom_loss(y_true, y_pred):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    return keras.losses.mean_absolute_error(y_true_f, y_pred_f)

def custom_loss2(y_true, y_pred):
    y_true_f = y_true[2,:]
    y_pred_f = y_pred[2,:]
    return keras.losses.mean_absolute_error(y_true_f, y_pred_f)
```

```python
## 결과를 플롯하기 위한 함수 
def comparison_plot(start=0, n_figure=200, show=False):
    plot_row = 2
    plot_col = 3
    #n_figure = 200

#     predict_img = model.predict([low_img_val, disp_val, se_val],  batch_size, verbose=2)

    # np.savetxt('predict.txt',y_predict)

    plt.rcParams["figure.figsize"] = (20,8)
    for i_figure in range(0, n_figure):

        # fig = plt.figure(nrows=4, ncols=5)
        fig = plt.figure()
        
        
        # plt.subplots_adjust(hspace = 0.5, wspace = 0.3)
        # print('predict start:', start + i_figure * plot_row*plot_col)

        start_time = time.time()
        for i_col in range(0, plot_col):
            for i_row in range(0, plot_row):

                index = start + i_figure * plot_row * plot_col  + plot_row * i_col + i_row
                # print('index',index)
                one_y_orig = high_img_test[index].reshape(img_rows*upscale, img_cols*upscale).transpose()
                one_y_low = low_img_test[index].reshape(img_rows, img_cols).transpose()
                one_y_predict = predict_img[index].reshape(img_rows*upscale, img_cols*upscale).transpose()


                plt.subplot(plot_col, plot_row * 3, (plot_row *3 * i_col + i_row*3) + 1) # 2 * 3 * 4 + 1 * 3 + 3 = 24 + 3 + 3 = 30
                plt.imshow(one_y_low, cmap='Greys', vmin=0, vmax=1)
                plt.xticks([])
                plt.yticks([])
                # plt.ylabel("n:%d" % index+1)
                if i_col == plot_col - 1:
                    plt.xlabel("low-res.")

                plt.subplot(plot_col, plot_row*3, (plot_row *3 * i_col + i_row*3) + 2)
                plt.imshow(one_y_predict, cmap='Greys', vmin=0, vmax=1)
                plt.xticks([])
                plt.yticks([])
                # plt.ylabel("n:%d" % index + 1)
                if i_col == plot_col - 1:
                    plt.xlabel("predicted")
                plt.subplot(plot_col, plot_row * 3, (plot_row *3 * i_col + i_row*3) + 3)
                plt.imshow(one_y_orig, cmap='Greys', vmin=0, vmax=1)
                plt.xticks([])
                plt.yticks([])
                # plt.ylabel("n:%d" % index + 1)
                if i_col == plot_col - 1:
                    plt.xlabel("high-res.")


        end_time= (time.time() - start_time)/plot_row
        # print('computing time for predict : %f s' % end_time)
        #plt.savefig("../train/vae/cmp_epoch%03d_latent%02d_i%04d.png" % (epochs, latent_dim, i_figure))
        plt.tight_layout()

        
        import os
        outDir = "./output_fig"
        if not os.path.isdir(outDir):
            os.mkdir(outDir)
            
        plt.savefig("./output_fig/out_figure_%s.png" % i_figure,dpi=300)
        if show==True:
#             plt.figure(figsize=(2,2)) 
#             plt.rcParams['figure.figsize'] = [100, 50]
            plt.show()
        plt.close()



```
