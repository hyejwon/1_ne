## 데이터를 augmentation 하기위한 함수 정의 (90도씩 돌리고 flip -> 8배)

```python
from scipy.ndimage import rotate
import math

# plt.imshow(low[500].reshape(img_rows, img_cols), cmap='Greys', vmin=0, vmax=1)
# plt.show()

def rot_augment(img):
    img90 = np.rot90(img,k=1,axes=(2,1))
    img180 = np.rot90(img,k=2,axes=(2,1))
    img270 = np.rot90(img,k=3,axes=(2,1))

    return np.concatenate((img, img90, img180, img270), axis=0)
    # return img

def flip_augment(img):
    fl=img[:,:,::-1,:]

#     fl=img[:,::-1,:,:]

    return np.concatenate((img,fl),axis=0)

def flip_augment_vector(img):
#     fl=-img[:,::-1,:,:]
    fl=-img[:,:,::-1,:]
    return np.concatenate((img,fl),axis=0)

def augment(img):
    return flip_augment(rot_augment(img))
    #return rot_augment(img)

def rot_vector(vx,vy,k):
    vx = np.rot90(vx,k,axes=(2,1))
    vy = np.rot90(vy,k,axes=(2,1))

    th = (2*3.141592)*(k*90/360)
    print('th=',th)
    vx_rot = vx*math.cos(th)-vy*math.sin(th)
    vy_rot = vx*math.sin(th)+vy*math.cos(th)
    
    return vx_rot, vy_rot

def rot_augment_vector(vx, vy):

    vx90, vy90 = rot_vector(vx,vy,1)
    vx180, vy180 = rot_vector(vx,vy,2)
    vx270, vy270 = rot_vector(vx,vy,3)

    vx = np.concatenate((vx,vx90,vx180,vx270),axis=0)
    vy = np.concatenate((vy,vy90,vy180,vy270),axis=0)
    
    return flip_augment_vector(vx), flip_augment(vy)

def rot_augment_vector2(v):

    vx = v[:,:,:,:3]
    vy = v[:,:,:,3:]
    
    vx_arg, vy_arg = rot_augment_vector(vx,vy)
    return np.concatenate((vx_arg,vy_arg),axis=3)


#return vx, vy
    


```

## 모델정의

```python
def SubpixelConv2D(input_shape, scale=4):
    """
    Keras layer to do subpixel convolution.
    NOTE: Tensorflow backend only. Uses tf.depth_to_space
    Ref:
        [1] Real-Time Single Image and Video Super-Resolution Using an Efficient Sub-Pixel Convolutional Neural Network
            Shi et Al.
            https://arxiv.org/abs/1609.05158
    :param input_shape: tensor shape, (batch, height, width, channel)
    :param scale: upsampling scale. Default=4
    :return:
    """
    # upsample using depth_to_space
    def subpixel_shape(input_shape):
        dims = [input_shape[0],
                input_shape[1] * scale,
                input_shape[2] * scale,
                int(input_shape[3] / (scale ** 2))]
        output_shape = tuple(dims)
        return output_shape

    def subpixel(x):
        return tf.depth_to_space(x, scale)


#     return Lambda(subpixel, output_shape=subpixel_shape, name='subpixel')
    return Lambda(subpixel, output_shape=subpixel_shape)
```

```python
interpolation='bilinear' #'nearest'   #upscaling시 방법 nearest는 같은 값으로 채우고, bilinear는 interpolation 합니다.
act_f = 'relu'
n_ker = int(64)
init_f = 'glorot_uniform' #'he_normal'
repeat = 4

## 입력 tensor의 정의 
if seInfo == True:
    n_ch = 3
else:
    n_ch = 2
    
    
low = Input(batch_shape=(None,) + input_size)
high = Input(batch_shape=(None,) + original_img_size)
disp = Input(batch_shape=(None,) + disp_size)

if permut_inv == True:  #
    diff = Input(batch_shape=(None,) + (12,12,3,n_ch))
else:
    diff = Input(batch_shape=(None,) + diff_size)

se = Input(batch_shape=(None,) + se_size)
se_sum = Input(batch_shape=(None,) + (12,12,1))


## model의 정의 
from keras.layers import Conv3D, MaxPooling3D, Permute, Add
from keras.layers.advanced_activations import PReLU

# disp는 nodal information이고 se는 element information이므로 차원이 다릅니다. (n,m),(n+1,m+1)
# 따라서 차원을 맞춰주기 위해서 padding이 없고 filter size 가 2인 cnn을 먹여서 차원을 같게 맞춰준후 concatenate합니다. 

def ConvBlock(fn, fs, act, pad, repeat,x):

    y = x
    for i in range(repeat):
        s = y
        y = Conv2D(fn,fs,padding = pad, kernel_initializer = init_f)(y)
        y = BatchNormalization()(y)
#         y = Activation(act)(y)
        y = PReLU()(y)
        y = Conv2D(fn,fs, padding = pad, kernel_initializer = init_f)(y)
        y = BatchNormalization()(y)
    
        y = keras.layers.add([s,y])
    
    y = Conv2D(fn,fs,padding = pad, kernel_initializer = init_f)(y)
    y = BatchNormalization()(y)
    y = keras.layers.add([x,y])
    return y


# def SrModel(strInfo=False, seInfo=False, permut_inv=False):
def SrModel():
    if strInfo==True:
        if permut_inv == True:
########################################################################################################
## structural information 처리 부분 
############################ new structure for structural information (puermutation invariation layer)
            conv0 = Conv3D(int(n_ker/4),(3,3,1), activation = act_f, padding = 'same', kernel_initializer = init_f)(diff)
            conv0 = Conv3D(int(n_ker/2),(3,3,1), activation = act_f, padding = 'same', kernel_initializer = init_f)(conv0)
            conv0 = Conv3D(int(n_ker),(3,3,1), activation = act_f, padding = 'same', kernel_initializer = init_f)(conv0)
    #         conv1 = Permute((1,2,4,3))(conv0)
            conv0 = MaxPooling3D((1,1,3))(conv0)
            conv0 = Permute((1,2,4,3))(conv0)
            conv0 = Reshape((12,12,int(n_ker)))(conv0)

        
##########################################################################################################
        else: ###### old structure (no permutaional invariance)
            conv0 = Conv2D(32,3,activation=act_f, padding = 'same', kernel_initializer = init_f)(diff)
            conv0 = Conv2D(64,3,activation=act_f, padding = 'same', kernel_initializer = init_f)(conv0)
            conv0 = Conv2D(128,3,activation=act_f, padding = 'same', kernel_initializer = init_f)(conv0)
        
##########################################################################################################    



## main network Resnet + subpixelCNN
    y = Conv2D(n_ker,9, padding='same',kernel_initializer=init_f)(low)
    if strInfo==True:
        y = keras.layers.concatenate([y,conv0], axis=3)
    y = Conv2D(n_ker,3, padding='same',kernel_initializer=init_f)(y)
    y = PReLU()(y)

    out = [0]*3
    for n in range(3): ##  한번데 두배씩 8배 upscaling
        y = ConvBlock(n_ker, 3, 'relu', 'same',  repeat, y)
    #     y = ConvBlock(n_ker, 3, 'relu', 'same',  1, y)
        y = Conv2D(n_ker*4,3,padding='same',kernel_initializer=init_f)(y)
        y = SubpixelConv2D((12*(2**n),12*(2**n),n_ker*4), scale=2)(y)
        y = PReLU()(y)
        out[n] = Conv2D(1, 9, activation='sigmoid',padding='same')(y)
        
#         out[n] = y

    out24 = out[0]
    out48 = out[1]
    out96 = out[2]
    
    
    if strInfo==True:
        if seInfo == True:
            x = [low,diff, se_sum]
        else:
            x = [low,diff]
    else:
        x = low
    if prog_loss == True:       
        model = Model(x,[out24, out48, out96])
    else:
        model = Model(x,out96)
    return model
```
